# Final output of this notebook is a heatmap [here](#Heatmap)

```python
import pandas as pd
tweets_df = pd.read_csv("earthquake/earthquake_tweets.csv")
tweets_df
```

```python
## Remove records that cannot be converted to datetime
m = pd.to_datetime(tweets_df['Datetime'], format='%Y-%m-%d', errors='coerce').isna()
tweets_df = tweets_df[~m]

tweets_df
```

```python
## Eliminate tweets which dont possess coordinates info

tweets_df = tweets_df.dropna(subset=['coordinates'])
tweets_df
```

```python
## Convert to datetime

tweets_df['Datetime'] = pd.to_datetime(tweets_df['Datetime'],format='%Y-%m-%d')
tweets_df
```

```python
## Handle different kind of strings to extract coordinates

import re
import json

def geo(s):
    if 'Coordinates(' in s:
        #coords = re.findall(r'(?<==)\w*\.?\w?\w?\w?\w?', str(s))
        coords = re.findall(r'(?<==)-?\w*\.?\w?\w?\w?\w?\w?', str(s))
        return [float(i) for i in reversed(coords)]
    
    else:
        coords = json.loads(s.replace("\'", "\""))['coordinates']
        return [float(i) for i in coords]

tweets_df['coordinates'] = tweets_df['coordinates'].apply(lambda x: geo(x))
tweets_df
```

```python
## Eliminate coordinates not having both longtitude and latitiude

tweets_df = tweets_df[~(tweets_df['coordinates'].apply(len)<1)]
tweets_df
```

```python
## Get latitude and longtitude

tweets_df['latitude'] = tweets_df['coordinates'].apply(lambda x: x[1])
tweets_df['longtitude'] = tweets_df['coordinates'].apply(lambda x: x[0])
tweets_df
```

```python
## Shows dates

tweets_df.groupby(tweets_df['Datetime'].dt.date)['coordinates'].apply(list).apply(len)
```
